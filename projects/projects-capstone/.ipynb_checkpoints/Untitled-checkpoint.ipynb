```python
outlist = []

factor = 8

with open('/Users/peidacai/Desktop/turnstile_160102.csv', 'r') as f:
    count = 0
    for line in f:
        
        if count % factor == 0:
            count += 1
            outlist.append(line)
        else:
            count += 1
            pass
f.close()
```

```python
with open('/Users/peidacai/Desktop/NYC Subway mini data/mini_turnstile_160604.csv', 'w') as f:
    
    f.writelines(outlist)
f.close()
```
