```python
# Functions to get listing details from scrapped website(s)

def get_listing_details(listing_placard):
    
    listing = []
        
    # Name
    try:
        
        for d in soup.findAll(class_='biz-name js-analytics-click'):
            listing.append(d.text.strip())
                 
    except:
        listing.append('')
    
    # Rating
    try:
        for d in soup.findAll(class_='biz-rating biz-rating-large clearfix'):
            listing.append(d.find('div')['title'].strip())

    except:
        listing.append('')
     
    # Number of reviews
    try:
        for d in soup.findAll(class_='biz-rating biz-rating-large clearfix'):
            listing.append(d.text.strip())
    except:
        listing.append('')
    
    # $ ratings
    try:
        for d in soup.findAll(class_='business-attribute price-range'):
            listing.append(d.text.strip())
    except:
        listing.append('')
    
    # Category of business
    try:
        for d in soup.findAll(class_='category-str-list'):
            listing.append(d.text.strip().split())
    except:
        listing.append('')
    
    # street address
    try:
        for d in soup.findAll(class_='secondary-attributes'):
            listing.append(d.find('address').text.strip())
    except:
        listing.append('')
    
    # Web address
    try:
        listi=[]
        for d in soup.findAll(class_='js-analytics-click'):

            listi.append(d['href'])
        
        listing.append('https://www.yelp.com/biz'+listi[0])

    except:
        listing.append('')
    
    return listing
```
