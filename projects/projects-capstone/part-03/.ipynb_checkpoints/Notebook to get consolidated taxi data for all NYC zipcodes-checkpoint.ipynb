# Create a reference dataframe for all NYC zipcode on taxi data

```python
import pandas as pd
import numpy as np
import pickle
```

```python
with open('/Users/peidacai/Desktop/taxi_df_dropoff_in_nyc_only.pkl', 'r') as f:
    taxi = pickle.load(f)
```

```python
taxi.head(1)
```

```python
taxi.columns
```

```python
drophour = pd.get_dummies(taxi['dropoff_hour'], prefix='dropoff_hour')
```

```python
drophour.shape
```

```python
dow = pd.get_dummies(taxi['day_of_week'], prefix='day')
```

```python
sub_taxi = pd.concat([taxi['zipcode'], drophour, dow], axis = 1)
```

```python
sub_taxi.head(1)
```

```python
taxi_zip = pd.pivot_table(sub_taxi, values=(list(dow.columns) + list(drophour.columns)), index=['zipcode'], aggfunc=sum)
```

```python
# Load in a NYC Zipcode list

nyc_zip = pd.read_csv('../part-02/assets/data/NYC_zipcodes/nyc_zip.txt', header=None)
nyc_zip.head()
```

```python
# Create a list of zipcodes which we do not have taxi info for.

nan_zip = []

for i in nyc_zip[1]:
    if i not in taxi_zip.index:
        nan_zip.append(i)
```

```python
len(nan_zip)
```

```python
estimation = {}

# defining the upper bound
max_zip = max(taxi_zip.index)

for i in nan_zip:
    
    ref_list = []
    # Number of neighboring zipcodes to consider
    total = 2

    count_up = 0
    new_zip = i
    
    # Getting 2 from higher zipcodes
    
    while (count_up < (total/2)) and new_zip <= max_zip:

        new_zip += 1
        if new_zip in taxi_zip.index:
            ref_list.append(new_zip)
            count_up += 1
            new_zip += 1

        else:
            new_zip += 1

    count_down = 0
    new_zip = i
    
    # 2 from lower zipcodes
    while count_down < (total - count_up) and new_zip > 10001:
        new_zip -= 1
        if new_zip in taxi_zip.index:
            ref_list.append(new_zip)
            count_up += 1
            new_zip -= 1

        else:
            new_zip -= 1
    
    estimation[i] = ref_list
```

```python
round(65.5, 0)
```

```python
max_zip
```

```python
estimation
```

```python
# Create an empty dictionary
vals = {}

for i in estimation:
    
    est_list = []
    
    for j in taxi_zip.columns:
    
        est_list.append(round(taxi_zip[taxi_zip.index.isin(estimation[i])][j].mean(),0))

    # Assigning the list as values to the corresponding key
    vals[i] = est_list

# Converting to dataframe
neighbour = pd.DataFrame(vals).T
neighbour.columns = taxi_zip.columns
neighbour.head()
```

```python
# Join to existing dataframe to complete taxi data for all NYC zipcodes
neighbour_taxi = pd.concat([neighbour, taxi_zip], axis = 0)
neighbour_taxi.shape
```

```python
neighbour_taxi.head()
```

```python
neighbour_taxi['dropoff_count'] = neighbour_taxi.iloc[:,0:7].sum(axis = 1)
```
