```python
def change_wr(x):
    try:
        op = float(x)
    except:
        try:
            y = x.split('-')
            op = (float(y[0]) + float(y[1]))/2
        except:
            op = np.nan
    return op
```

```python
dfc['world_rank'] = dfc['world_rank'].apply(change_wr)
```

```python
dfc['world_rank'].unique()
```

```python
dfc['country'].unique()
```

```python
dfc['country'] = dfc['country'].apply(lambda x: "United States of America" if "Unisted" in x else x)
```

```python
dfc['country'] = dfc['country'].apply(lambda x: "United Kingdom" if "Unted" in x else x)
```

```python
pd.DataFrame(dfc['country'].unique())
```

```python
dfc['international'].unique()
```

```python
dfc['international'] = dfc['international'].apply(lambda x: np.nan if "-" in x else x)
```

```python
dfc['income'].unique()
```

```python
dfc.info()
```

```python
def clean_d(x):
    if x == '-':
        x = np.nan
    x = float(x)
    return x
```

```python

dfc[['international','income', 'total_score']] = \
dfc[['international','income', 'total_score']].applymap(clean_d)
```

```python
dfc.head()
```

```python
def clean_st(x):
    if type(x) == type(0.1):
        return x
    else:
        
        if len(x) >= 4:
            y = x.split(',')
            op = (float(y[0]) *10000) + float(y[1])
            return op
        else:
            op = float(x)
            return op
```

```python
dfc['num_students'] = dfc['num_students'].apply(clean_st)
```

```python
def clean_int(x):
    if type(x) != type(0.1):
        x = float(x[:-1])
    
    x = x/100
    return x
    
```

```python
dfc['international_students'] = dfc['international_students'].apply(clean_int)
```

```python
dfc.head()
```

```python
def clean_fmr(x):
    x = str(x)
    if ':' in x:
        x = x.split(':')
        if x[0] == '100':
            output = 1.
        else:
            output = float(x[0])/float(x[1])
        return output
    else:
        try:
            output = float(x)
            return output
        except:
            output = np.nan
            return output
    
```

```python
dfc['female_male_ratio'] = dfc['female_male_ratio'].apply(clean_fmr)
```

```python
df = dfc.dropna()
```

```python
dfx = df[['world_rank', 'teaching', 'international', 'research', 'citations', 'income',\
               'num_students', 'student_staff_ratio', 'international_students',\
              'female_male_ratio']]
dfx.head()
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.cross_validation import train_test_split

X = dfx
y = df['total_score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33)

lm = linear_model.LinearRegression()
lm.fit(X_train, y_train)

pred = lm.predict(X_test)

mse = mean_squared_error(y_test, pred)
print "MSE:", mse
print "Mean coefficient:", abs(lm.coef_).mean()
```
