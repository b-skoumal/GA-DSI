```python
# Create a histogram by creating a list comprehension + a function to turn the data into inches
```

```python
def ht_inch(ht):
    if ht == 0:
        ht_in = 0
    elif ht == "Under 5 ft.":
        ht_in = 59
    elif ht == '6\'6" and above':
        ht_in = 73
    else:
        ht_split = ht.split("'")
        ht_in = (float(ht_split[0]) * 12) + float((ht_split[1].split('"'))[0])
        
    return ht_in
```

```python
char[['Height']] = char[['Height']].fillna(0)
char[['Height']]
```

```python
char_ht = char[['Height']].dropna()
char_ht
```

```python
char_ht = char_ht['Height'].apply(ht_inch)
char_ht
```

```python
import seaborn as sns

bw = 80
fig = char_ht.hist(figsize=(12,6), color='hotpink', bins = bw)

plt.style.use('fivethirtyeight')

# get the current axis
ax = plt.gca()

ax.set_ylabel('Frequency', fontsize=14)
ax.set_xlabel('Height in inches', fontsize=14)
ax.set_title('Height of air travel passengers', y=1.01)
```
